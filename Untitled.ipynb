```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataframe = pd.read_csv("Zomato_data.csv")
print(dataframe)
```

```python
dataframe
```

convert the data type colum rate - rate

```python
def handleRate(value):
      value=str(value).split('/')
      value=value[0];
      return float(value);

dataframe['rate']=dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe.info()
```

```python
dataframe.head()
```

Conclusion:- Majority of the restaurent fall in dining category

```python
sns.countplot(x=dataframe['listed_in(type)'], palette="Set2")  # You can try different palettes like "viridis", "coolwarm", etc.
plt.xlabel("Type of Restaurant", color="red", size=20)
plt.ylabel("Count", color="red", size=20)
plt.title("Count of Types of Restaurants", size=24)
plt.xticks(rotation=45)
plt.show()
```

Conclusion:- Majority of the restaurent fall in dining category

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})

plt.plot(result.index, result['votes'], c="green", marker="o")
plt.xlabel("Types of Restaurant", color="red", size=20)
plt.ylabel("Votes", color="red", size=20)
plt.title("Votes by Type of Restaurant", size=24)
plt.show()
```

Conclution:- dining Restaurent has recieved maximum votes .

```python
counts, bins, patches = plt.hist(dataframe['rate'], bins=5)
for i, patch in enumerate(patches):
    if i % 2 == 0:
        patch.set_facecolor("skyblue") 
    else:
        patch.set_facecolor("orange")

plt.title("Rating Distributions")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()
```

The majority of restaurant recieved the rating 3.5 to 4.0
